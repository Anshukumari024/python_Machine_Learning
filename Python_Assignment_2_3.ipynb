```python
import numpy as np
import time
A = np.random.rand(100, 100)
B = np.random.rand(100, 100)
start_time = time.time()
result = np.dot(A, B)

try:
    det = np.linalg.det(result)
    print("Determinant:", det)

    if np.isclose(det, 0):
        print("Matrix is singular and not invertible.")
        inv = None
    else:
        inv = np.linalg.inv(result)
        print("Inverse calculated successfully.")

except np.linalg.LinAlgError as e:
    print("Matrix operation failed:", e)
    inv = None

end_time = time.time()
print("Time taken: {:.4f} seconds".format(end_time - start_time))
```

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}
df = pd.DataFrame(data)
df['Passed'] = df['Marks'] >= 80

print("Name and Marks columns:")
print(df[['Name', 'Marks']])

print("Students with Marks > 80:")
print(df[df['Marks'] > 80])

highestmarks = df['Marks'].max()
toppers = df[df['Marks'] == highestmarks]
print("Student with the Highest Marks:")
print(toppers)
```

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}
df = pd.DataFrame(data)

df.loc[1, 'Marks'] = None
df.loc[4, 'Age'] = None

print("Original DataFrame with Missing Values:")
print(df)

print("Missing Values (True indicates missing):")
print(df.isnull())

mean_marks = df['Marks'].mean()
df['Marks'] = df['Marks'].fillna(mean_marks)


print("DataFrame after Filling Missing Marks with Mean (%.2f):" % mean_marks)
print(df)

df_cleaned = df.dropna(subset=['Age'])

print("DataFrame after Dropping Rows with Missing Age:")
print(df_cleaned)
```

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

groupedmean = df.groupby('Gender')[['Age', 'Marks']].mean()
print("Mean Age and Marks by Gender:")
print(groupedmean)

gendercount = df['Gender'].value_counts()
print("Number of Students in Each Gender Group:")
print(gendercount)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')  
print("First 5 rows of the dataset:")
print(df.head())
print("Dataset Info:")
print(df.info())

print("Summary Statistics:")
print(df.describe())

print("Class Distribution (Species):")
print(df['species'].value_counts())
print("Missing Values:")
print(df.isnull().sum())

df.hist(figsize=(10, 6))
plt.suptitle("Histograms of Iris Features")
plt.show()

sns.boxplot(data=df)
plt.title("Boxplot of all Features")
plt.show()

sns.pairplot(df, hue='species')
plt.suptitle("Pairplot by Species", y=1.02)
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
